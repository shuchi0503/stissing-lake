```python
import pandas as pd
```

```python
## Data imported from Division of Water Monitoring Data Portal by DEC - /
# https://experience.arcgis.com/experience/301748017d7d40649bc5082fc1c5365e#data_s=id%3AdataSource_1-194997771f3-layer-13%3A189%2Cid%3AdataSource_1-194997753f2-layer-12%3A2386449%2Cid%3AdataSource_1-1949977d66f-layer-14%3A17383

df= pd.read_csv("stissing-water-quality-raw-data.csv")
df
```

```python
df.info()
```

```python
## separating year from date time and checking if 2024 year data is included 

df["EVENT_DATETIME"] = pd.to_datetime(df["EVENT_DATETIME"], errors="coerce")

df["Year"] = df["EVENT_DATETIME"].dt.year
df.query("Year == 2024")     
```

```python
## just columns i need 

df.head()[["Year", "PARAMETER_NAME", "RESULT_CATEGORY", "RESULT_VALUE", "UNIT", "RESULT_QUALIFIER", "RESULT_QUALIFIER_DESCRIPTION"]]
```

```python
df["Year"].value_counts()
```

```python
##filter dataset for the 3 parameters i need and calling it df_parameters

df_parameters = df.query(
    "(PARAMETER_NAME == 'phosphorus') or (PARAMETER_NAME == 'chlorophyll-a') or (PARAMETER_NAME == 'secchi_disk_depth')"
)[
    ["Year", "PARAMETER_NAME", "RESULT_CATEGORY", "RESULT_VALUE", "UNIT", "RESULT_QUALIFIER", "RESULT_QUALIFIER_DESCRIPTION"]
]
df_parameters
```

```python
## remove columns you don't need for the graph 
## store filtered data into a new dataframe

df_stissing = df.query(
    "(PARAMETER_NAME == 'phosphorus') or (PARAMETER_NAME == 'chlorophyll-a') or (PARAMETER_NAME == 'secchi_disk_depth')"
)[
    ["Year", "PARAMETER_NAME", "RESULT_VALUE", "UNIT"]
]

df_stissing
```

```python
df_stissing.to_csv("stissing-lake-graph.csv", index=True)
```

```python
df_chloro = df_stissing.query("PARAMETER_NAME == 'chlorophyll-a'")[["Year", "PARAMETER_NAME", "RESULT_VALUE", "UNIT"]]
df_chloro
```

```python
##Let's calculate median result for each year to make it easier to put into a graph (there are diff no of samples for diff years)

## Chlorophyll median

df_chloro_median = df_chloro.groupby('Year').agg({
    'RESULT_VALUE': 'median',
    'UNIT': 'first'  # Since unit should be the same for all chlorophyll-a measurements
}).reset_index()

df_chloro_median

# Rename the column for clarity
df_chloro_median = df_chloro_median.rename(columns={'RESULT_VALUE': 'MEDIAN_RESULT_VALUE'})

df_chloro_median
```

```python
df_chloro_median.to_csv("stissing-chlorophyll-median.csv", index=True)
```

```python
df_phos = df_stissing.query("PARAMETER_NAME == 'phosphorus'")[["Year", "PARAMETER_NAME", "RESULT_VALUE", "UNIT"]]
df_phos
```

```python
## Phosphorus median

df_phos_median = df_phos.groupby('Year').agg({
    'RESULT_VALUE': 'median',
    'UNIT': 'first'
}).reset_index()
# Rename the column for clarity
df_phos_median = df_phos_median.rename(columns={'RESULT_VALUE': 'MEDIAN_RESULT_VALUE'})

df_phos_median
```

```python
df_phos_median.to_csv("stissing-phos-median.csv", index=True)
```

```python
## in the phos- median table, we can see that the levels have been slowly increasing since 2021. I seems int he 2024 report, 
# they used median as 0.0195, so they are using 2007-2017 data. 
```

```python
df_sec = df_stissing.query("PARAMETER_NAME == 'secchi_disk_depth'")[["Year", "PARAMETER_NAME", "RESULT_VALUE", "UNIT"]]
df_sec
```

```python
#### Secchi disk depth median - for water clarity

df_sec_median = df_sec.groupby('Year').agg({
    'RESULT_VALUE': 'median',
    'UNIT': 'first'
}).reset_index()
# Rename the column for clarity
df_sec_median = df_sec_median.rename(columns={'RESULT_VALUE': 'MEDIAN_RESULT_VALUE'})

df_sec_median 
```

```python
df_sec_median.to_csv("stissing-sechhi-median.csv",index=True)
```

```python
## Now calculate the percent change for each year for all three parameters - this might help put all three in 1 graph, otherwise it's 
# difficult because they have different units

df_chloro_median['PERCENT_CHANGE'] = df_chloro_median['MEDIAN_RESULT_VALUE'].pct_change() * 100

df_chloro_median

```

```python
df_phos_median['PERCENT_CHANGE'] = df_phos_median['MEDIAN_RESULT_VALUE'].pct_change() * 100

df_phos_median
```

```python
df_sec_median['PERCENT_CHANGE'] = df_sec_median['MEDIAN_RESULT_VALUE'].pct_change() * 100

df_sec_median
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df_chloro_median['Year'], df_chloro_median['PERCENT_CHANGE'], marker='o', label='Chlorophyll-a')
plt.plot(df_phos_median['Year'], df_phos_median['PERCENT_CHANGE'], marker='s', label='Phosphorus')
plt.plot(df_sec_median['Year'], df_sec_median['PERCENT_CHANGE'], marker='^', label='Secchi Disk Depth')

plt.xlabel('Year')
plt.ylabel('Percent Change from Baseline (%)')
plt.title('Water Quality Parameter Trends (Percent Change from 2019)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Stack all three into long format
df_chloro_median['Parameter'] = 'Chlorophyll-a'
df_phos_median['Parameter'] = 'Phosphorus' 
df_sec_median['Parameter'] = 'Secchi Disk Depth'

stissing_percent_change = pd.concat([
    df_chloro_median[['Year', 'Parameter', 'PERCENT_CHANGE']],
    df_phos_median[['Year', 'Parameter', 'PERCENT_CHANGE']],
    df_sec_median[['Year', 'Parameter', 'PERCENT_CHANGE']]
], ignore_index=True)

stissing_percent_change
```

```python
stissing_percent_change.to_csv("stissing-percent-change.csv", index=True)
```
